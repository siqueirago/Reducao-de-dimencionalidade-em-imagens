# **Implementação em Python para transformar uma imagem colorida em níveis de cinza e binarizada**

```python
from IPython import get_ipython
from IPython.display import display
import cv2
from google.colab.patches import cv2_imshow
```

```python
def converter_para_cinza(imagem_colorida):


  return cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

def binarizar_imagem(imagem_cinza, limiar):

  _, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)
  return imagem_binarizada

# Carrega a imagem colorida
imagem_colorida = cv2.imread("/content/drive/MyDrive/imagens/cachorro.jpg")

# Converte para níveis de cinza
imagem_cinza = converter_para_cinza(imagem_colorida)

# Define o limiar para a binarização
limiar = 128  # Você pode ajustar este valor

# Binariza a imagem
imagem_binarizada = binarizar_imagem(imagem_cinza, limiar)

# Exibe as imagens using cv2_imshow() instead of cv2.imshow()
cv2_imshow(imagem_colorida)
cv2_imshow(imagem_cinza)
cv2_imshow(imagem_binarizada)
cv2.waitKey(0)
cv2.destroyAllWindows()

# cv2.waitKey(0) and cv2.destroyAllWindows() are not needed with cv2_imshow()
#cv2.waitKey(0)
#cv2.destroyAllWindows()

# Salva as imagens (opcional)
# cv2.imwrite() expects a filename as the first argument
# Providing the image data as the first argument will result in an error
# Please provide valid filenames for saving the images
cv2.imwrite("imagem_cinza.jpg", imagem_cinza)
cv2.imwrite("imagem_binarizada.jpg", imagem_binarizada)
```
